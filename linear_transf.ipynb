```python
import numpy as np
import matplotlib.pyplot as plt

# Define a uniformly spaced vector of x1 and x2 values
x1 = np.arange(0, 5.25, 0.25)  # 0:0.25:5 in MATLAB
x2 = x1

# Get all possible coordinate combinations
X1, X2 = np.meshgrid(x1, x2)

# Visualize the original grid
plt.figure(figsize=(8, 8))
plt.plot(X1, X2, 'b.', label='Original Points')
```

```python
# Linear transformation 1: Change the scaling
A = np.array([[0, 2], [2, 0]])
X = np.vstack([X1.ravel(), X2.ravel()])
Y = A @ X
plt.plot(Y[0, :], Y[1, :], 'm.', label='Transformation 1')
```

```python
# Linear transformation 2: Stretch out the grid
B = np.array([[0, 1], [1, 5]])
Y = B @ X
plt.plot(Y[0, :], Y[1, :], 'c.', label='Transformation 2')
```

```python
# Linear transformation 3: Stretch out the grid a bit more
C = np.array([[0, 1], [1, 50]])
Y = C @ X
plt.plot(Y[0, :], Y[1, :], 'r.', label='Transformation 3')

# Show plot with legends
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear Transformations')
plt.legend()
plt.axis('equal')
plt.grid()
plt.show()
```
